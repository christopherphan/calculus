# Average cost and marginal cost illustration

This is an illustration I made for my Applied Calculus lecture on average cost.
It's supposed to illustrate two concepts:
* The critical point of the average cost function will occur when average cost is equal to marginal cost.
* The average cost function is minimized when the slope of the line through the origin and the graph of the cost function is minimized.

The code to move the axes is from the SciPy Lecture Notes ```matplotlib``` tutorial, by Nicolas Rougier, Mike Müller, and Gaël Varoquaux, at <http://www.scipy-lectures.org/intro/matplotlib/index.html#moving-spines>, and is used under a Creative Commons Attribution 4.0 International License.

````python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

avgcost = lambda costfunc : (lambda q: costfunc(q)/q)
margcost = lambda costfunc : (lambda q: (costfunc(q+0.001) - costfunc(q))/0.001)
````

````python
cost = lambda q: 0.000017*q**3 - 0.025*q**2 + 14.5*q + 5000
fig1 = plt.figure(1, figsize=(10, 7))
ax1 = fig1.add_subplot(211)
xvals = np.arange(0.01, 1600, 0.01)
yvals1 = cost(xvals)
plt.plot(xvals, yvals1, color="red",  linewidth=2.5, label="Total Cost")
plt.grid(True)

slopeline = lambda q: margcost(cost)(912.073)*(q - 912.073) + cost(912.073)
yvals3 = slopeline(xvals)
plt.plot(xvals, yvals3, color="green", linewidth=2.5, linestyle=":")
 
# Following is from: http://www.scipy-lectures.org/intro/matplotlib/index.html#moving-spines
ax = plt.gca()  # gca stands for 'get current axis'
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))

plt.xlabel("Quantity (items)")
plt.ylabel("Cost (dollars)")
plt.ylim(0, cost(1600))
plt.legend(loc='upper left')

ax2 = fig1.add_subplot(212)
yvals2 = avgcost(cost)(xvals)
yvals3 = margcost(cost)(xvals)
plt.ylim(0, 60)
plt.plot(xvals, yvals2, color="magenta",  linewidth=2.5, label="Average Cost")
plt.plot(xvals, yvals3, color="orange",  linewidth=2.5, label="Marginal Cost")
plt.grid(True)

# Following is from: http://www.scipy-lectures.org/intro/matplotlib/index.html#moving-spines
ax = plt.gca()  # gca stands for 'get current axis'
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))

plt.xlabel("Quantity (items)")
plt.ylabel("Cost (dollars/item)")

plt.legend(loc='upper right')

fig1.savefig("2016-11-03_1.png")
````
